# I am going to define a function for this plot so that I can apply to all the 9 plots 

```python
def histplot(step):
    rg='rg'+str(step)
    unmod='unmod'+str(step)
    fig, ax = plt.subplots(figsize=(10,10))
    df[rg]=df[rg].map(lambda x: x+360 if x <-50 else x)
    df[unmod]=df[unmod].map(lambda x: x+360 if x <-50 else x)
    sns.distplot(df[rg],color='blue',label=rg)
    sns.distplot(df[unmod],color='green',label=unmod)
    plt.legend()
    sns.set_context("notebook", font_scale=2, rc={"lines.linewidth": 2.5})
    ax.set_xlabel(rg +" and " + unmod + " gamma torsion in degrees")
    plt.savefig("gamma_hist_"+str(step)+".png")
```

## let's try it on two steps

```python
histplot(69)
```

```python
for i in range(69,78):
    print i
    histplot(i)
```
